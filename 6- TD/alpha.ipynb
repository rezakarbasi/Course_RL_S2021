```python
import numpy as np

alpha = 0.01

x = np.arange(1000)
# original = np.sin(x/100)
original = (x**2 - 100*x -100)
original[original>150] = 150
noisy = original + np.random.randn(1000)*500
```

```python
filtered = np.zeros(1000)

for i in range(1,len(original)):
  filtered[i] = filtered[i-1] + alpha * (noisy[i]-filtered[i-1])
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.subplot(221)
plt.plot(original)
plt.title('original signal')

plt.subplot(222)
plt.plot(noisy)
plt.title('noisy signal')

plt.subplot(223)
plt.plot(filtered)
plt.title('filtered signal')


plt.subplot(224)
plt.plot(original)
plt.plot(filtered)
plt.title('filtered vs original signal')
```

```python
def plot(alpha ,original ,std):
  lenn = len(original)
  noisy = original + np.random.randn(lenn)*std
  filtered = np.zeros(lenn)

  for i in range(1,len(original)):
    filtered[i] = filtered[i-1] + alpha * (noisy[i]-filtered[i-1])

  import matplotlib.pyplot as plt

  plt.figure()

  plt.subplot(221)
  plt.plot(original)
  plt.title('original signal')

  plt.subplot(222)
  plt.plot(noisy)
  plt.title('noisy signal')

  plt.subplot(223)
  plt.plot(filtered)
  plt.title('filtered signal')


  plt.subplot(224)
  plt.plot(original)
  plt.plot(filtered)
  plt.title('filtered vs original signal')

  plt.show()
```

```python
x = np.arange(1000)
original = np.sin(x/100)

plot(0.1,original,1)
```
