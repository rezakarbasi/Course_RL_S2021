```python
import gym
 
env = gym.make("FrozenLake-v0",map_name="8x8")
env.render()
```

```python
import numpy as np

np.random.seed(1)
policy = np.random.rand(64,4)*4
policy = np.exp(policy)
policy /= np.sum(policy,axis=1)[:,None]
print(policy)
```

```python
def step(action):
  new_state, reward, done, info = env.step(action)
  if done:
    reward = -1
  if new_state==63:
    reward = 1
  return new_state, reward, done, info
```

```python
v=np.zeros(64)
gamma = 1
alpha = 0.1

for i in range(10000):
  done = False
  state = env.reset()
  while done != True:
    action = np.random.choice(range(4),p=policy[state])
    new_state, reward, done, info = step(action)
    v[state] = v[state]+alpha*(reward+gamma*v[new_state]-v[state])
    state = new_state

  # env.render()
```
